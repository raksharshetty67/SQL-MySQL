````SQL
CREATE TABLE table_name1 (
    column1 INT,
    column2 VARCHAR(50),
    column3 INT
);

INSERT INTO table_name1 (column1, column2, column3) VALUES
(1, 'A', 100),
(2, 'B', 200),
(3, 'C', 300),

-- duplicates
(1, 'A', 100),
(2, 'B', 200),

-- another unique
(4, 'D', 400),

-- duplicate again
(3, 'C', 300),
(3, 'C', 300);

````
---------------------------------
### **1. Write a SQL Queries to Remove Duplicate Rows in a Table.**
#### **Method 1:**
````SQL
WITH cte AS (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY column1, column2, column3 ORDER BY column1) AS rn
    FROM table_name
)
DELETE FROM cte
WHERE rn > 1;
````
--------------------------------

````SQL
#### **Method 2:**
DELETE t1
FROM table_name t1
JOIN table_name t2
ON t1.column1 = t2.column1
AND t1.column2 = t2.column2
AND t1.column3 = t2.column3
AND t1.rowid > t2.rowid;

````
