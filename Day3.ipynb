### **1. Find cumulative salary for each employee from January to July.**
````SQL
SELECT Emp_id, Month, Salary,
       SUM(Salary) OVER (
           PARTITION BY Emp_id 
           ORDER BY 
               CASE Month
                   WHEN 'January' THEN 1
                   WHEN 'February' THEN 2
                   WHEN 'March' THEN 3
                   WHEN 'April' THEN 4
                   WHEN 'May' THEN 5
                   WHEN 'June' THEN 6
                   WHEN 'July' THEN 7
               END
       ) AS Cumulative_Salary
FROM table_name2;
````
